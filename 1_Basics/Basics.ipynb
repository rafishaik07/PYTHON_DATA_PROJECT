# String Notations

 String slicing in Python is a technique used to extract a portion or a substring from a larger string.

Here's a breakdown of how it works:

Syntax: **string[start:stop:step]**

string: The original string you want to slice.

start: (Optional) The index of the first character to include in the slice. If omitted, it defaults to the beginning of the string (index 0).

stop: (Optional) The index of the character after the last character to include in the slice. The character at this index is not included. If omitted, it defaults to the end of the string.

step: (Optional) The step or increment to use when slicing. If omitted, it defaults to 1 (meaning consecutive characters). A negative step can be used to slice in reverse.

**Examples**:

**string[start:stop]**: Extracts characters from start up to (but not including) stop.
string[:stop]: Extracts characters from the beginning up to (but not including) stop.
string[start:]: Extracts characters from start to the end of the string.
string[:]: Creates a copy of the entire string.
string[::step]: Extracts characters with a specified step size.
string[::-1]: Reverses the string.

```python
class LukesList:
    def __init__(self):
        """Initialize an empty list."""
        self._items = []

    def add(self, item):
        """Add an item to the end of the list."""
        self._items.append(item)

    def __getitem__(self, index):
        """Retrieve an item by index."""
        return self._items[index]

    def __setitem__(self, index, value):
        """Set an item at a specific index."""
        self._items[index] = value

    def __repr__(self):
        """Return a string representation of the list."""
        return str(self._items)

    def __len__(self):
        """Return the length of the list."""
        return len(self._items)
```

```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate=0.1, symbol = "$"):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol
    self.total_salary = (base_salary)*(1+bonus_rate)
    self.bonus_rate = (self.total_salary - base_salary) / (base_salary)

  def __repr__(self):
    return f'{self.symbol}{self.base_salary:,.0f}'

  def calculate_salary(self):
    return f'{self.symbol}{self.total_salary:,.0f}'

  def calculate_bonus(self):
    return f'{self.symbol}{self.bonus_rate:,.0f}'

```
