# copy()

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()

df_original
```

```python
df_altered = df_original
```

```python
df_altered['salary_year_avg'].head(10)
```

```python
df_altered.loc[:10, 'salary_year_avg']
```

```python
median_salary_year = df_altered['salary_year_avg'].median()
```

```python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary_year)
```

```python
df_altered['salary_year_avg']
```

```python
df_original['salary_year_avg']
```

```python
id(df_original) == id(df_altered)
```

```python
print('The ID of df_original is:', id(df_original))
print('The ID of df_altered is:', id(df_altered))
print('Is df_original the same as df_altered?', id(df_original) == id(df_altered))
```

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()

df_original
```

```python
df_altered = df_original.copy()
```

```python
print('The ID of the original DataFrame is:', id(df_original))
print('The ID of the altered DataFrame is:', id(df_altered))
print('Are the two DataFrames the same?', id(df_original) == id(df_altered))
```

```python
median_salary_year = df_altered['salary_year_avg'].median()
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary_year)
```
