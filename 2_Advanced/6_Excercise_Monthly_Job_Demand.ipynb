# Excercise - Analyzing Data Science Job Demand

```python
# importing the necessary libraries
import pandas as pd
from datasets import load_dataset

# loading the dataset
dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()

# Data Cleaning

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
```

```python
df_US
```

```python
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')

df_US
```

```python
df_US.pivot_table(index='job_posted_month',columns='job_title_short', aggfunc='size')
```

```python
df_US['job_posted_month_num'] = df_US['job_posted_date'].dt.month
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month',columns='job_title_short', aggfunc='size')

```

```python
df_US_pivot.reset_index(inplace=True)
df_US_pivot['month_num'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
df_US_pivot = df_US_pivot.sort_values(by='month_num', inplace=True)
df_US_pivot.set_index('job_posted_month', inplace=True)
df_US_pivot.drop(columns='month_num', inplace=True)
```

```python
df_US_pivot
```

```python
df_US_pivot.plot(kind='line')
```

```python
top_3 = df_US['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()

top_3
```

```python
df_US_pivot[top_3].plot(kind='line')
```

```python
import matplotlib.pyplot as plt

df_US_pivot[top_3].plot(kind='line')
plt.title('Top 3 Job Titles Posted in the US by Month')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
